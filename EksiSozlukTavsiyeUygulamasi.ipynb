```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import requests

headers = {"User-Agent" : "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.67 Safari/537.36"}
url_list = ["https://eksisozluk.com/sozluk-yazarlarindan-tavsiyeler--3540401?p="]


# Retrieving the data from URL's.
pre_data = []
def get_data(url_link):
    for x in range(1,26): #25*10 entry
        page = requests.get(url_link+str(x),headers=headers)
        soup = BeautifulSoup(page.content,"html.parser")
        for entry in soup.find_all("div",{"class" : "content"}):
            pre_data.append(entry.get_text())
            
for url in url_list:
    get_data(url)
```

```python
eksi=pre_data[:250]
```

# Jaccard Similarity

```python
#Text mining: find the similarity between twotext documents using the number of terms used in both documents

class eksiyasamkocum():
    
    def __init__(self, veri):
        self.veri = veri
        
    def benzerlik(self, A, B):
        """
        Jaccard benzerligi
        """
        A = [a.lower() for a in A.split()]
        B = [b.lower() for b in B.split()]

        A_kesisim_B = {a for a in A if a in B}
        A_kesisim_B_sayisi = len(A_kesisim_B)
        A_birlesim_B_sayisi = len(A) + len(B) - len(A_kesisim_B)
        return A_kesisim_B_sayisi / A_birlesim_B_sayisi
    

    def dinle_ve_tavsiye_ver(self):

      ##  Kullanicidan girdi olarak bir cevap al:
       
      ##    - cevabın verimizdeki sozler ile benzerligini hesapla ve en yakin sozu dondur
   
            
        cevap = input("Merhaba, \nHangi konuda tavsiye istersin?(Hayat/Sevgi/Gayret/Idare)\n\t Cevabiniz: ").lower()
       
        if cevap.isdigit():
             print(f'Sayı değeri girmeyiniz !')
        else:
            f = lambda x: self.benzerlik(x, cevap)
            sonuc = self.veri.Sozler.apply(f)
            # sonuc_idx = sonuc.sort_values(ascending=False).index[0]
            return self.veri.iloc[sonuc.idxmax()].Sozler
```

```python
koc = eksiyasamkocum(veri)
```
